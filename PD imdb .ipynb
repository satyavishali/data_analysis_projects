# 22. Find Unique Values From Genre 


```python
data.columns
```

```python
data['Genre']
```

```python
list1=[]

for i in data['Genre']:
    v=i.split(',')
    list1.append(v)

```

```python
list1
```

```python
one_d=[]

for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique=[]

for i in one_d:
    if i not in unique:
        unique.append(i)
```
